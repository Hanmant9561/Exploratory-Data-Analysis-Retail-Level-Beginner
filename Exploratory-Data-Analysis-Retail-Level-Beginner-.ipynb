# Relationship Analysis

```python
corelation = data.corr()
sns.heatmap(corelation,xticklabels=corelation.columns, yticklabels=corelation.columns,annot=True,cmap='Blues')
```

```python
sns.pairplot(data)
```

```python
sns.distplot(data['Sales'])
```

```python
sns.distplot(data['Profit'])
```

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=data)
```

```python
data.groupby('Sub-Category')['Profit','Sales'].agg(sum).plot.bar()
plt.title('Profit and Sales across Sub-Categories')
```

From the above graph,the Profit of Copiers are high,and the Sales of Chairs & Phones are high among the others.

```python
data['Cost'] = data['Sales'] - data['Profit']
data['Profit %'] = (data['Profit']/data['Cost'])/100
```

```python
Profit_data = data.sort_values(['Profit %','Sub-Category'], ascending= False), data.groupby('Profit %')
```

```python
plt.rcParams['figure.figsize'] = (10,5)
sns.distplot(data['Sales'],kde = False)
plt.title("Distribution of Sales")
plt.xlabel("Range of sales")
plt.ylabel("No. of Sales Count")
plt.show()
```
